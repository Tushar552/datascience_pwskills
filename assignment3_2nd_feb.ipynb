### Q1. Explain with an example each when to use a for loop and a while loop.

#### For loop:
A for loop is a control flow statement that executes code repeatedly for a particular number of iterations. In this control flow statement, the keyword used is for. The for loop is used when the number of iterations is already known.



```python
# Example of for loop

l = [1, 2, 3,4]  # Initialization
for i in l:  # Condition and Updation
    print(i)
```

#### While loop:
A loop that executes a single statement or a group of statements for the given true condition. The keyword used to represent this loop is "while". A "while" loop is used when the number of iterations is unknown. The statement repeats itself till the boolean value becomes false. In a while loop, the condition is tested at the start, also known as the pre-test loop


```python
# Example of while loop

i = 1  # Initialization
while i < 4:  # Condition
    print(i)
    i += 1  # Updation
```

### Q2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

##### sum of the first 10 natural numbers using For loop.

```python
sum1 = 0
for i in range(1,11):
    sum1= sum1+i
print(f"sum = {sum1}")
```

##### product of the first 10 natural numbers using For loop.

```python
product = 1
for i in range(1,11):
    product= product*i
print(f"product = {product}")
```

##### sum of the first 10 natural numbers using While loop.

```python
sum1 = 0
i = 0
while i <= 10:
    sum1 = sum1 + i
    i = i+1
print(f"sum = {sum1}")
```

##### product of the first 10 natural numbers using While loop.

```python
product = 1
i = 1
while i <= 10:
    product = product * i
    i = i+1
print(f"product = {product}")
```

### Q3. Create a python program to compute the electricity bill for a household.

#### The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.

```python
unit = int(input("enter units = "))
total_charge = 0
i = 1
while i <= unit:
    if i <= 100:
        total_charge = total_charge + 4.5
    if i > 100 and i <= 200:
        total_charge = total_charge + 6
    if i > 200 and i <= 300:
        total_charge = total_charge + 10
    if i > 300:
        total_charge = total_charge + 20
        
    i = i+1
print(f"Total Electricty Bill Charge = {total_charge} rupees")
        
```

### Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print that list.

##### Using For loop

```python
l = list(range(1,101))
l1 = []
for i in l:
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        l1.append(i)
print(l1)
```

##### Using While loop

```python
l = list(range(1,101))
l1 = []
i = 1
while i <= len(l):
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        l1.append(i)
    i = i+1
print(l1)
```

### Q5. Write a program to filter count vowels in the below-given string.
#### string = "I want to become a data scientist"

```python
string = "I want to become a data scientist"
vowel = ['a','A','e','E','i','I','o','O','u','U']
vowel_string = []
vowel_count = 0
for i in string:
    for j in vowel:
        if i == j:
            vowel_count = vowel_count + 1
            vowel_string.append(i)
print(f"vowel_count = {vowel_count}")
print(f"Vowels in given string = {vowel_string}")

    
```
